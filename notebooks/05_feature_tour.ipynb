## 2 — Flux bubble (interactive)

Open `01_flux_bubble_explorer.ipynb` for full widgets; quick static snapshot here:

```python
%matplotlib inline
import matplotlib.pyplot as plt
from hfb.bubble.warp_conduit import flux_bubble_metric
from hfb.utils.grid import cartesian_grid
x, y = cartesian_grid(64, 64, extent=3.0)
m = flux_bubble_metric(x, y, dx=float(x[0,1]-x[0,0]))
plt.imshow(m['omega'], origin='lower'); plt.title('Conformal Ω'); plt.colorbar(); plt.show()
```
