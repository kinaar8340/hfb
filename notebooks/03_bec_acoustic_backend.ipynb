# BEC Acoustic Backend

Thomas-Fermi density, imprinted vortex rings, and BEC acoustic metrics.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider, FloatSlider

from hfb.bec.acoustic import bec_acoustic_metric
from hfb.bec.bogoliubov import bogoliubov_dispersion
from hfb.bec.vortex import vortex_ring_velocity
from hfb.utils.grid import cartesian_grid

x, y = cartesian_grid(96, 96, extent=3.0)
```

```python
def explore_bec(num_vortices=8, ring_radius=1.0, n0=1.0):
    vx, vy, phase = vortex_ring_velocity(x, y, ring_radius=ring_radius, num_vortices=num_vortices)
    metric = bec_acoustic_metric(x, y, vx, vy, n0=n0, radius=1.5)

    fig, axes = plt.subplots(1, 3, figsize=(13, 4))
    axes[0].imshow(metric["density"], origin="lower")
    axes[0].set_title("Thomas-Fermi density")
    axes[1].imshow(metric["c_s"], origin="lower")
    axes[1].set_title("Sound speed c_s")
    axes[2].imshow(metric["ergoregion"], origin="lower")
    axes[2].set_title(f"Ergoregion (frac={metric['ergoregion'].mean():.3f})")
    plt.tight_layout()
    plt.show()

    k = np.linspace(0.01, 8.0, 100)
    eps = bogoliubov_dispersion(k, c_s=1.0)
    plt.figure(figsize=(5, 3))
    plt.plot(k, eps)
    plt.xlabel("k")
    plt.ylabel("ε(k)")
    plt.title("Bogoliubov dispersion (ℏ=1)")
    plt.show()

interact(
    explore_bec,
    num_vortices=IntSlider(2, 16, 1, value=8, description="vortices"),
    ring_radius=FloatSlider(0.4, 1.5, 0.05, value=1.0, description="ring_R"),
    n0=FloatSlider(0.5, 2.0, 0.1, value=1.0, description="n0"),
)
```
