# SLM Vortex Ring Export

Export flux-bubble vortex conduit phase holograms. Couples to `vqc_proto` when available via `VQC_PROTO_PATH`.

```python
%matplotlib inline
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider, FloatSlider

from hfb.integration.vqc_proto import vqc_proto_available
from hfb.optics.slm_export import (
    SLMExportConfig,
    VortexConduitSpec,
    export_flux_bubble_hologram,
    flux_bubble_vortex_field,
    field_to_slm_phase,
)

print("vqc_proto coupled:", vqc_proto_available())
OUT = Path("../outputs/notebook_slm")
```

```python
def preview_and_export(num_vortices=12, ring_radius=1.2, winding=1):
    cfg = SLMExportConfig.from_vqc_preset("generic_512")
    spec = VortexConduitSpec(
        num_vortices=num_vortices,
        ring_radius_mm=ring_radius,
        winding=winding,
    )
    field = flux_bubble_vortex_field(spec, cfg)
    phase = field_to_slm_phase(field)

    fig, axes = plt.subplots(1, 2, figsize=(10, 4))
    axes[0].imshow(np.abs(field), origin="lower")
    axes[0].set_title("|E| — vortex ring superposition")
    axes[1].imshow(phase, origin="lower", cmap="twilight", vmin=0, vmax=2*3.14159)
    axes[1].set_title("SLM phase (rad)")
    plt.tight_layout()
    plt.show()

    summary = export_flux_bubble_hologram(OUT, spec=spec, cfg=cfg)
    print(summary)

interact(
    preview_and_export,
    num_vortices=IntSlider(4, 24, 1, value=12, description="vortices"),
    ring_radius=FloatSlider(0.5, 2.0, 0.05, value=1.2, description="ring_mm"),
    winding=IntSlider(1, 3, 1, value=1, description="winding"),
)
```
