# Acoustic Metrics (Symbolic)

SymPy symbolic line elements, horizons, and conformal Ricci scalars.

Requires: `pip install -e ".[symbolic]"`

```python
import sympy as sp
from sympy import init_printing
init_printing()

from hfb.analog_gravity.symbolic import (
    acoustic_line_element,
    acoustic_metric_inverse,
    horizon_condition,
    draining_vortex_velocity,
    evaluate_metric_numeric,
)

x, y = sp.symbols("x y")
vx, vy = draining_vortex_velocity(x, y, circulation=1.0, drain=0.1)
print("Draining vortex vx =")
display(vx)
```

```python
ds2 = acoustic_line_element()
horizon = sp.factor(horizon_condition())

print("Line element ds² =")
display(ds2)
print("Horizon condition (factored):")
display(horizon)
print("g^{μν} at v_x=0.3, v_y=0.4, c_s=1:")
display(sp.Matrix(evaluate_metric_numeric(0.3, 0.4, 1.0)))
```
